```python
%pylab inline

import pandas as pd
import seaborn as sns
import networkx as nx
```

```python
books  = pd.read_csv('./data/books.csv')
ratings = pd.read_csv('./data/ratings.csv')
```

```python
books.head()
```

```python
ratings.head()
```

```python
book_ratings = nx.DiGraph()
```

```python
nx.from_pandas_edgelist(df=ratings.iloc[:,:],
                        source='book_id',
                        target='user_id',
                        edge_attr='rating',
                        create_using=book_ratings)
```

```python
sns.set(style='white',
        rc={
            'figure.figsize':(8,8)
        })

nx.draw(book_ratings);
```
